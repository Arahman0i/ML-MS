```python
def calculate_subject_percentage(class_name: str, subject_name: str, obtained_marks: float, total_marks: float) -> float:
    """
    Calculates the percentage of marks obtained in a specific subject for a given class.

    Args:
        class_name (str): The name or identifier of the class (e.g., "10th Grade", "Class A").
        subject_name (str): The name of the subject (e.g., "Mathematics", "Physics").
        obtained_marks (float): The marks obtained by the student in the subject.
        total_marks (float): The maximum possible marks for the subject.

    Returns:
        float: The calculated percentage. Returns 0.0 if total_marks is 0 to avoid division by zero.
    """
    if total_marks == 0:
        print(f"Warning: Total marks for {subject_name} in {class_name} is 0. Cannot calculate percentage.")
        return 0.0

    percentage = (obtained_marks / total_marks) * 100
    return percentage

# --- Example Usage ---

# Define details for the jth class, ith subject, and kth marks
# For demonstration, let's assume:
# jth class = "10th Grade"
# ith subject = "Science"
# kth marks = 75 (obtained marks) out of 100 (total marks)

# Define variables for our specific scenario
class_id = "10th Grade"
subject_id = "Science"
student_obtained_marks = 75.5
subject_total_marks = 100

# Calculate the percentage using the function
calculated_percentage = calculate_subject_percentage(
    class_name=class_id,
    subject_name=subject_id,
    obtained_marks=student_obtained_marks,
    total_marks=subject_total_marks
)

# Print the results in a structured and well-documented way
print("--- Percentage Calculation Report ---")
print(f"Class: {class_id}")
print(f"Subject: {subject_id}")
print(f"Marks Obtained: {student_obtained_marks}")
print(f"Total Marks: {subject_total_marks}")
print(f"Calculated Percentage: {calculated_percentage:.2f}%")
print("-----------------------------------")

# Another example: Mathematics in 12th Grade
class_id_2 = "12th Grade"
subject_id_2 = "Mathematics"
student_obtained_marks_2 = 88
subject_total_marks_2 = 120

calculated_percentage_2 = calculate_subject_percentage(
    class_name=class_id_2,
    subject_name=subject_id_2,
    obtained_marks=student_obtained_marks_2,
    total_marks=subject_total_marks_2
)

print("\n--- Another Example ---")
print(f"Class: {class_id_2}")
print(f"Subject: {subject_id_2}")
print(f"Marks Obtained: {student_obtained_marks_2}")
print(f"Total Marks: {subject_total_marks_2}")
print(f"Calculated Percentage: {calculated_percentage_2:.2f}%")
print("-----------------------")

# Example with zero total marks
class_id_3 = "8th Grade"
subject_id_3 = "Art"
student_obtained_marks_3 = 25
subject_total_marks_3 = 0

calculated_percentage_3 = calculate_subject_percentage(
    class_name=class_id_3,
    subject_name=subject_id_3,
    obtained_marks=student_obtained_marks_3,
    total_marks=subject_total_marks_3
)

print("\n--- Example with Zero Total Marks ---")
print(f"Class: {class_id_3}")
print(f"Subject: {subject_id_3}")
print(f"Marks Obtained: {student_obtained_marks_3}")
print(f"Total Marks: {subject_total_marks_3}")
print(f"Calculated Percentage: {calculated_percentage_3:.2f}%")
print("-----------------------------------")
```

```python
"""
Module: MarksheetGenerator
Description: Generates a text-based representation of a marksheet.
Safety: This script operates entirely in-memory with no external dependencies
or data transmission.
"""

def generate_marksheet(student_name, roll_no, marks_data):
    """
    Generates a formatted text marksheet.

    :param student_name: str, name of the student
    :param roll_no: str, roll number
    :param marks_data: dict, keys as subject names, values as a list [theory, practical]
    """

    # SAFETY CHECK: Validate that the input data is not empty
    if not marks_data:
        raise ValueError("Error: Marks data cannot be empty.")

    print(f"--- CENTRAL BOARD OF SECONDARY EDUCATION ---")
    print(f"NAME: {student_name} | ROLL NO: {roll_no}")
    print("-" * 65)
    print(f"{'SUBJECT':<20} | {'TH':<4} | {'PR':<4} | {'TOTAL':<5}")
    print("-" * 65)

    grand_total = 0
    for subject, scores in marks_data.items():
        # Perform calculation
        total = sum(scores)
        grand_total += total

        # Displaying with basic formatting symbols
        print(f"{subject:<20} | {scores[0]:<4} | {scores[1]:<4} | {total:<5}")

    print("-" * 65)
    print(f"RESULT: PASS | GRAND TOTAL: {grand_total}")
    print("-" * 65)

# Example Usage
# Ensure names and IDs are handled safely as strings
try:
    student_record = {
        "ENGLISH CORE": [57, 0],    # [Theory, Practical]
        "MATHEMATICS": [83, 0],
        "PHYSICS": [57, 30],
        "CHEMISTRY": [63, 30],
        "COMPUTER SCIENCE": [60, 29]
    }

    generate_marksheet("ANKUR WARIKOO", "6231413", student_record)
except Exception as e:
    # Proper error documentation for accidental execution failures
    print(f"An unexpected error occurred during generation: {e}")
```

## Data frame

A **DataFrame** is a 2-dimensional, spreadsheet-like data structure provided by the `pandas` library. It organizes data into rows and columns, similar to a SQL table or an Excel sheet.

### Key Characteristics:
* **Tabular Structure:** Rows represent individual records, and columns represent distinct variables/features.
* **Heterogeneous Types:** Different columns can hold different data types (e.g., integers, strings, floats, dates).
* **Labeled Axes:** Both rows and columns have explicit labels (indexes), making data selection fast and intuitive.


```python
import pandas as pd # lib for data manipulation using data frames
df = pd.DataFrame() # this creates an empty data frame.

data = {
    "Element" : ['C', 'O', 'Fe', 'Mg', 'Xe'],
    "Atomic Number" : [6, 8, 26, 12, 54],
    "Atomic Mass" : [12, 16, 56, 24, 131]
}

df = pd.DataFrame(data)

df = df.set_index("Element")

df["Atomic Mass"] # displays the output in a much cleaner formatted table  than print function
```

```python
df["Energy (eV)"] = [5.47, 5.14, 0.12, 4.34, 7.01]
df["Energy (eV)"]
```

```python
# Print a row from the DataFrame
# The df.loc[index] function to print the entry "C"
print(df.loc["C"])
df.loc["Fe"]
```

```python
# Modifying and Creating Columns
# easily perform math across entire columns instantly without writing loops.
df["Neutrons"] = df["Atomic Mass"] - df["Atomic Number"]

df.loc["Fe"]
```

```python
# Removes a column (axis=1 means column)
df = df.drop("Neutrons", axis=1)
df.loc["Fe"]
```
