```python
import os
from dotenv import load_dotenv

# Load the keys from the local hidden .env file
load_dotenv()

# Retrieve the secure variable from memory
api_key = os.getenv("GEMINI_API_KEY")

# Verify it works without exposing the key text
if api_key:
    print("API key loaded successfully into memory.")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=api_key)

text= "Suggest three novel cathodes for Al ion batteries."

response = model.invoke(text)
print(response.content)
```
